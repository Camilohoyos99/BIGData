```python
import pyspark
import pandas as pd
```

```python
sc = pyspark.SparkContext(appName="ejemplo")

rdd = sc.textFile("input/NASDAQsample.csv")

resultado = rdd.map(lambda x: x.split(",")[7]).map(lambda x: float(x)).takeOrdered(10, key=lambda x: -x)

resultado
```

```python
df=pd.read_csv("input/NASDAQsample.csv",header=None)
df.head(5)
```

```python
# Ordenar el DataFrame por la columna 7 en orden descendente y seleccionar las 10 primeras filas
top10 = df.sort_values(by=7, ascending=False).head(10)

# Seleccionar las columnas 2, 3 y 7 del DataFrame filtrado
resultado = top10.iloc[:, [1, 2, 7]]
```
